## Introduction
- 一個component 建立在class 元素階層當中，且該component 不像extension method, 它可以存取class 裡面non-common/specific 的 component
- 但Visitor 只是比較觀念性的只會去呼叫visit() 方法，真正運作的機制是透過Double Dispatch來實踐
    - Dispatch 就是派工也就是後續要call 哪一個function
    - double dispatch : 在call function 前的依據，=> 會看 request's name, visitor's type and 被拜訪的element type, 再去invoke 相對應的方法。
