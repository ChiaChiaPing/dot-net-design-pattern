### 2. Implementation
- 由於我們只想讓物件被創建一次，所以當我們在設計時我們部會允許外部以間接或是直接的方式去創建，而是只提供存取的權限，然而該物件的創建則會交由給該物件所屬的類別去執行。
    - lazy : 延遲物件創建，作用是提升城市執行效能，只用到現階段所佔用的記憶體空間。例如Database 連接，當我建立一個Dictionary 要來存放db 資料時，在載入之前我不會站用到空間，載入後才會佔用到，目的是提升程式執行的效率。若沒有lazy，則在仔入之前就會一直佔用空間了。

## 另外一個與Singleton pattern 作用類似的 Pattern MonoState
- MonoState ，將全部的field 都設為private 且 static ，只提供 prop 給外界存取，
    - 由於外界不知道class 實踐方式，所以當然就認為是指向同一個物件，但其實這是由於static緣故，讓我們都是指向同一個物件 ，
    - 容易實踐，不過效能效率比 singleton pattern 差且難維護，
    - 我覺得他並不是 Singleton ，因為他是創建不同的物件，只是裡面的屬性都指向同樣的地方(Static 緣故)，讓我們有種即使創建不同物件，這些不同物件都指向同一個物件的假象。

### 三原則 實踐 SingleTon object
- 在 class 建構 private ctor
- 用該 private ctor 建構物件
- 建立 public prop 供外部去存取該物件
    - 並搭配使用Factory 方法
