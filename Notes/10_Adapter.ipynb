### Adapter  - Structural Pattern
- 代表將 現有的 interface 去 conform to 需求的interface
- 就很像是一個轉接器 把 interface/class 為 A的 轉換成 interface/class 為B 的
    - 這樣的你就可以把介面跟物件連想在一起， 該轉接器就是在做轉換物件或是介面
    - 所以當你在設計某個方法要做不同型別之間的轉型時，該方法的簽章 就是 Adapter 一種 case。

### defect of Adapter and how to resolve ( reduce Adapter Caching )
- Defect : Generate temporary Information when Adaptering
- Solution : 用一個Dictionaries 的形式 與 HashCode 來存放所有線 他們產生的所有的 points collections.
    - Generate Equality Members 幫助我們判斷兩物件是否為一樣

### Generic Value Adapter
- 透過 Factory Methods 可以去創建 int, 或是 float 的 n 維度向量
    - 採用 recursive generic : 代表我針對我定義的泛型下 where 限制型別時，下與該泛型一樣的型式
    - 在下 where condition 時可以搭配 new () 代表 可以利用該泛型創建物件。
    - 任何 class 可以透過 static 與 operator 建立運算子，以 a+b 為例，在呼叫時，當錢的物件已經佔據兩個物件之中的一個，故剩下一個與被加物件一樣型別的物件做運算，且回傳的型別需與做運算的物件之型別一致。
    - static class 內建立 non-static 類別時，可以直接透過 static class 去存取 該 non-staticc class
    - 範例放在專案內。
