#### 2.1 Event keyword
- like delegate method, but a little bit different
- 帶入被下 event 的 variable assigned 的方法簽章除了EvenHandler<T> 定義之泛型的參數之外，還必須在方法簽章的第一個位置設立 object sender
- 該event 在invoke 時也必須要帶入 this 以及 T 所定義型別的參數
    - 若沒有參數 assigned 的function 也要有EventArgs 的型別參數，只是在invoke 時帶入 EventArgs.Empty
