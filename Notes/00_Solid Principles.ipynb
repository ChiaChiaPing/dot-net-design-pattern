### 1. .NET Design Pattern Introductions
- Design Pattern 只是一個架構性的方法，並且特別適用於 object-oriented-programming languages，提升程式的易讀性，以及可靠性
    - 這裡的 design pattern 是指 object-oriented design pattern, not the functional pattern
- 共有四類 design patterns

## 2. Solid Design Principles
- SOLID 是由五種 design principle 所組合起來的，由每一種principle 的名字開頭而成的 design principles

### 2.1 - Single Responsibility Principles  ( 單一方法 )
- One class 只專注在class自己本身應該要做的事情，不應該實作太多與之無關或是應該交由其他的 class 去做的事情（方法）
- 一個 class 都只有專屬的職責在那邊，一個類別只有一個改變的原因
- 難處：
    - class 命名
    - 及命名後怎麼將所有的適當地分給所有的class（職責難以分清）
    - 且顆粒度難控制，代表有時候職責會分太過細，導致有很多class ，維護不易
    - 若有很多職責會同時變動，那就將這些歸類的一個 class

### 2.2 - Open-Closed Principles
- 比較正確的定義應該是：software entries (class,module,function) can open for extension, close for modification
    - 代表可以再擴充功能的同時又不改變原先的code
- 代表藉由繼承class或是實作interface 來擴充功能，但是又能夠做到close modification，不改變原先class裡面的內容
    - like extension method 的做法
- 當方法定義 Ienumeratble 為回傳值時，可以善用 <b>yield return</b> 代表我return 但是我們繼續做該發法的內容，直到array 裡面的元素均完成了任務
- 當我們實作具有泛型的介面是，實作該介面的class 也可以無需指定泛型為何，可延用泛型，提升開發的彈性，

### 2.3 - Liskov Subsituition Principles
- 就是一個概念，當子類別的方法有 override時，若我們將某物件向上轉型成父類別，並呼叫該方法，則呼叫的內容會是 子類別在此方法定義的內容，而不是父類別的內容，這在繼承當中最為常見。
    - 藉由 override 與 virtual 來實踐
- 若方法當中欲執行的內容很簡單，只有一行code的話，可以採取匿名函數的做法來實踐
- 可以 sequence-派職
    - v1 = v2 = v3 ...
- nameof(variable) 可以轉成 variable 的字串
    - 利用 \$ 與 "{}"可以 將variable 直接嵌入在字串裡面

### 2.4 Interface Segerations (分離分割) Principles
- 不應該強迫用戶去依賴他們未使用的方法
    - like 實作介面，但只用某部分的方法，其他的方法根本沒用到，此時就違反 ISP
    - 若要解決的話，就可以從介面的角度變成是功能性的角度去切入，根據各個小小的功能性去建立介面。

### 2.5 Dependencies Inversion Principles
- High level Module not should depend on Low Level Module, both should depend on abstractions 抽象不應該依賴在細節，但細節應該依賴在抽象
    - 基本上可以透過介面與抽象類別的方式，將會共同使用的資訊給抽象畫，而細節則是交給繼承的子類別來實作
- List<T> T 可以是一個 tuple，並且可在做迭代的時候 透過 list.where 搭配存取 tuple內的某些item 來篩選物件
    - 前提是要 using  由 NuGet Package Manager 的 valueTuple 的套件才可以使用
    - using System.Linq 與 System.Ccollections.Generic
- 實作介面的方法時，方法的型別一定要依樣，即使是具有繼承估線的類別也不行
