### use IEnumerable{T} , 當要再方法或是屬性回傳 iteratorable list 時 可以這樣用, 
    - 回傳的型別會是一個list 並且elemeny 的 type 會是 T
    - 並且 利用 yield return 去決定該回傳的list 應該放什麼樣的元素
    - 善用 linq functional programming 去做後續的操作
### 另外一種實踐 iterator design pattern 的方式 先做 IEnumerable 取得 迭代氣 Enumerator（collection） 再用 Enumerator 去取得 collection 裡面的 element


```python
import pandas as pd
import numpy as np
```

```python
class Node():
    def __init__(self,value=0,left=None,right=None):
        self.value = value
        self.left = left
        self.right = right
```

```python
def traverse(root):
    print(root.value)
    if root.left:
        traverse(root.left)
    if root.right:
        traverse(root.right)
```

```python
a = Node(1)
b = Node(2)
c = Node(3)
a.left = b
a.right = c
```

```python
traverse(a)
```
