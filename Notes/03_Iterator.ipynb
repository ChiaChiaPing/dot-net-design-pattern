```python
import pandas as pd
import numpy as np
```

```python
class Node():
    def __init__(self,value=0,left=None,right=None):
        self.value = value
        self.left = left
        self.right = right
```

```python
def traverse(root):
    print(root.value)
    if root.left:
        traverse(root.left)
    if root.right:
        traverse(root.right)
```

```python
a = Node(1)
b = Node(2)
c = Node(3)
a.left = b
a.right = c
```

```python
traverse(a)
```
