### Command
- 可以利用一個物件下 Command 的概念，like GUI 介面再做操作時用物件來執行該action
    - 可以執行一個 serialize actiobn 也可以執行 undo 等 action
    - 可以把物件視為是command ，切這物件裡面可以有一個變數去存放執行這個command 的結果 (ex: boolean value)

### Composite Command
- Composite Patter combine Command Pattern
    - 很多小的command 包成一個一大組的Compisite Command ，然後只執行一次
    - 一個class如果繼承 Ienumerable<T> like List<T>, 就代表這個class他本身就是一個可以存放 T 型別的迭代物件
        - 就刻已把該類別創建的物件視為是一個集合物件，並且去做集合操作 或是 functional programming
