### Gamma Categorization
- Design Pattern 基於Solid Design Principle 上分成三個 Design Patterns 我們統稱Gamma Categorization
    - Creational Patterns : (related to creation of object) 的 patterns
    - Structural  : 與 sttructure of object (class member ) 有關的patterns
    - Behaviourial : no cebtral theme ( 不像上述兩者有特定的核心 )，因情況的不同有不同的patterns
    

### Creational Patterns - Builder Pattenr Introduction
- Problems : 當我們要去創建物件，我們可以直接去 call 建構子，但是若該建構子參數很多，若還是以同樣的方式去呼叫的話，會很沒效率
- 另一類似的 problem : 或是很複雜的 task 可以定義於 class 的建構子，外部只要呼叫該建構子帶入少許參數就可以建立了
- solution : 我們可以藉由 piece-by-piece 的方式去建立物件，like 小物件慢慢建立，最後建立成大物件
    - 實踐方式 : 透過 Builder 的 pattern的方式 幫助我們實踐以解決我們遇到的問題。
        - like StringBuilder 幫助我們實現，另可以piece-by-piece 的方式去建立Html element 樹狀的結構，在外部只要 call 相對應的 Constructor(Builder)，就可以幫助我們建立我們想要的物件了（HTML element 樹狀的結構）
        - fluent Interface(流式接口) : 像是 stringbuilder 的壹些方法，他呼叫這些方法的回傳型別與他呼叫前本身相同，所以可以一直去呼叫，雃這樣的形式就是 fluent interface

### Fluent Inherited builder
- 若要用繼承的方式去實踐 fluent interface，若沒有適當的設計方式，會很容易出錯，像是若呼叫父類別所定義的方法，則該回傳的型別則為負類別，這樣就不能去call 子類別的其他方法，在這樣的情況，我們只能要求該物件從最後的子類別依序網上呼叫方法才可行，所以問題很多，而至於怎麼解決，可以透過 recursive generic 的方式去解決。

###  Fluent Faceted(刻面，小面) Builder
- 代表我欲生成的物件過程中，我可以定義在Builder，去執行相對應的task，而處理這些Task 可以由不同的小Builder完成，且這些builder 內定義的方法"回傳值"都是自己，比如說可以指派reference 物件的不同屬性，然後於大Builder 最後回傳同一個referene 物件，讓我們可以實踐 fleuent Builder
    - 同一個 reference 物件其實like string builder 的概念有點像
- static(固定寫法) implicit operator A(B b)
    - 建立一個功能支援 B 隱性轉型的功能。
