### Introduction
- Template Method - High Level

- 抽象類別裡面可以加入建構子
    - 雖然不能再創建，但可以在利用子別創建物件時 因會呼叫父親的預設建構子而被呼叫到
    - 可以想像成是設計一個當任一子類別要創建時若都要呼叫的方法, 就可以把該方法視為是 父親抽象類別的建構子

## Two Kinds
- Template Method
- Functional Template Method
    - 善用delegate method, 只訂定一個大流程，而這個大流程裡面細部做的各個項目（方法）可以由外部來實踐
    - ex Func<bool> or Action
