## Factories
- 與 object 創建的 pattern 有關

### 1. Abstract Factories
- 我可以透過一個方法去實際創建<b>由實體的物件向上癮性轉型成介面</b>的物件
- 或是說我把創建戊介的方法定義在界面，並用一個類別去實作該就面裡面的方法，透過該方法去創建物件
    - ForExample : Tea/Coffee Factory 實作 IHotDrinkFactory 介面的方法，向上轉型，透過單一的窗口，去創建不同類型的物件。
    - IHotDrinkFactory 就是抽象工廠的範例
- HotDrinkMachine
    - Available drink, IHotDrinkFactory {Prepare()} 
        - IHotDrink : Consume()
- Ps. Activator : 動態的方式建立物件


### 2. Abstract Factories and OCP 
- List<裡面可以塞Ｔuple>
    - 且 取Tuple 裡面的內容值可以透過Item 去取直
- 補充說明：
    - c.IsAssignFrom(t) : 代表 t 是否可以轉型成c
    - c.Assembly.GetTypes() ：取得c 他所屬的assembly 底下所有定義的型別
        - 透過這樣的方式我們就可以在外部去做擴充，且不用修改原本的程式
            - 因為 我們預先設定好的 code 會去取得 assembly 底下所有的reference type，所以自然而然，我們在同一個Assembly 所新增的class也會跟被存取到
        - class, enum,interface
    - 要針對type去做 operation時，記得要將你欲操作的class type前面加入 typeof

### 3. Conclusion
- example 當你不想讓外部建立物件時，可以將負責建立物件的工廠類別作為該物件所屬類別裡面的 內部類別
